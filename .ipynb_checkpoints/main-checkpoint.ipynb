```python
#imports
import pandas as pd
import numpy as np
import cv2
from sklearn.metrics import confusion_matrix
import random
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
#instance of class holds the info for oe subject
class Subject:
    def __init__(self):
        self.image_list = []
        
    def read_in_images(self, subject_num, image_len):
        image_list = []
        for img_index in range(0, image_len):
            path = "dataset/s" + str(subject_num + 1) + "/" + str(img_index + 1) + ".pgm"
            next_image = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
            image_list.append(next_image)
        #this is the lists of images for that subject
        self.image_list = image_list
```

```python
#instance of class holds every subject in the imageset
class Imageset:
    def __init__(self):
        #attributes of image set
        #list of subjects
        self.subject_list = []
        #number of images each subject has
        self.images_per_subject = -1
        #array of flattened images
        self.image_array = np.zeros((0))
        #array that holds the centered data
        self.centered_array = np.zeros((0))
        #array that holds the output of the pca
        self.cut_pca_array = np.zeros((0))
        #explained variance from pca
        self.explained_variance = []
        #pca instance
        self.pca = -1
        self.basic_knn_accuracy = -1
        self.cross_validated_accuracy = -1
        self.pca_x = np.zeros((0))
        self.pca_y = np.zeros((0))
        
    #main read-in method
    def read_in_subjects(self, subject_len, image_len):
        subject_list = []
        for subject_index in range(0, subject_len):
            next_subject = Subject()
            next_subject.read_in_images(subject_index, image_len)
            subject_list.append(next_subject)
        #list of subjects in the imageset
        self.subject_list = subject_list
        self.images_per_subject = image_len
    
    #gets n random subject indexes
    def get_random_image_indexes(self, n):
        random_list = []
        for each in range(n):
            while True:
                #get a random
                next_random = random.randint(0, self.images_per_subject - 1)
                #if random number isn't already one of the randomly generated indexes
                if next_random not in random_list:
                    #add to the list, break out of current while loop
                    random_list.append(next_random)
                    break
        #return the list
        return random_list
    
    #reduces the current images
    def resize_images(self, original_images, resize_rate):
        subject_list = []
        for sub in original_images.subject_list:
            resized_sub = Subject()
            for img in sub.image_list:
                #take each diminension multiplied by the resize rate
                new_width = int(img.shape[0]*resize_rate)
                new_height = int(img.shape[1]*resize_rate)
                resized_img = cv2.resize(img, (new_width, new_height))
                resized_sub.image_list.append(resized_img)
            subject_list.append(resized_sub)
        self.subject_list = subject_list
        self.images_per_subject = original_images.images_per_subject
        return

    #use a random subset of n images from each subject of the the orig_imageset  
    def get_random_subset(self, orig_imageset, n): 
        #get the random indexes
        rand_list = orig_imageset.get_random_image_indexes(n)
        subject_list = []
        #for each subject in the original imageset,
        for original_subject in orig_imageset.subject_list:
            #make a new subject
            next_subject = Subject()
            #for each random index,
            for rand_num in rand_list:
                #add that subject's image index to the new subject
                next_subject.image_list.append(original_subject.image_list[rand_num])
            subject_list.append(next_subject)
        self.subject_list = subject_list
        self.images_per_subject = 5
            
    #flatten each image and make each image a row in a 2D numpy array
    def get_image_array(self):
        row_num = len(self.subject_list)*self.images_per_subject
        col_num = self.subject_list[0].image_list[0].shape[0]*self.subject_list[0].image_list[0].shape[1]
        image_array = np.zeros((row_num, col_num))
        for sub_index, sub in enumerate(self.subject_list):
            for img_index, img in enumerate(sub.image_list):
                image_array[(sub_index*self.images_per_subject)+img_index] = img.flatten()
        self.image_array = image_array
        
    #get the mean of each column, and subtract the mean from each value
    def centerize_array(self):
        mean_array = np.mean(self.image_array, axis = 0)
        centered_array = np.zeros((self.image_array.shape[0], self.image_array.shape[1]))
        for col_index, col in enumerate(mean_array):
            centered_array[:,col_index] = rand_images.image_array[:,col_index] - col
        self.centered_array = centered_array
    
    #get an array where each index corresponds to the subject index of each image
    def get_y_array(self):
        y_list = []
        for index in range(len(self.subject_list)):
            for img_count in range(self.images_per_subject):
                y_list.append(index)
        return np.asarray(y_list)
            
```

```python
#function not connected to a class

#do a pca on the data, get the explained variance
def perform_pca(data_array, num_of_components):
    pca = PCA(n_components = num_of_components)
    pca_array = pca.fit_transform(data_array)
    return pca, pca_array

#plot the output of the 2 column pca
def plot_pca(pca_array):
    #plot the eigenvalues
    plt.scatter(pca_array[:,0], pca_array[:,1])
    plt.show()
    
#find the last index that make up the top 90% of the explained variance
def get_last_index(variance_array):
    total_variance = 0
    for index, x in enumerate(np.nditer(variance_array)): 
        total_variance = total_variance + x
        if total_variance >= 0.9:
            return index
def cut_down_pca_array(pca_array, variance_array):
    last_index = get_last_index(variance_array)
    return pca_array[:,:last_index+1]

def perform_knn(x_train, x_test, y_train):
    #create the classifoer
    knn = KNeighborsClassifier(n_neighbors=1)
    #fit the classifier
    knn.fit(x_train, y_train)
    #predict the test set
    y_pred = knn.predict(x_test)
    #return the actual and predicted results of the test set
    return y_pred

def perform_cross_validation(x, y, num_folds):
    
    #combine the x and y's temporarily
    def combine_x_y(x, y):
        #make a new numpy array which is both x and y as columns
        total_array = np.zeros((x.shape[0], x.shape[1] + 1))
        total_array[:,:-1] = x
        total_array[:,-1] = y
        return total_array
    
    #split the data into equally-sized folds
    def split_dataset(total_array, num_folds):
        #shuffle the array
        np.random.shuffle(total_array)
        #now split it up into 5 different groups
        group_list = []
        rows_per_group = int(total_array.shape[0] / num_folds)
        for fold in range(num_folds):
            next_group = np.zeros((rows_per_group, total_array.shape[1]))
            next_group = total_array[fold*rows_per_group : ((fold+1)*rows_per_group)]
            group_list.append(next_group)
        return group_list
    
    def get_training_set(groups, test_index):
        train_groups = groups.copy()
        del train_groups[test_index]
        train_set = np.concatenate(train_groups, axis=0)
        return train_set
    
    def resplit_x_y(total_array):
        x = total_array[:,:-1]
        y = total_array[:,-1]
        return x, y
    
    total_array = combine_x_y(x, y)
    train_list = []
    group_list = split_dataset(total_array, num_folds)
    cross_validation_list = []
    for test_index, test_set in enumerate(group_list):
        next_train_set = get_training_set(group_list, test_index)
        x_train, y_train = resplit_x_y(next_train_set)
        x_test, y_test = resplit_x_y(test_set)
        y_pred = perform_knn(x_train, x_test, y_train)
        cross_validation_list.append((y_test, y_pred))
    return cross_validation_list

def print_accuracy_by_subject(y_actual, y_pred):
    results_df = pd.DataFrame({ "y_actual" : y_actual, "y_pred" : y_pred })
    actual_unique = results_df["y_actual"].unique()
    predict_unique = results_df["y_pred"].unique()
    unique_list = list(set(actual_unique) | set(predict_unique)) 
    unique_list.sort()
    print("accuracy by subject:")
    for value in unique_list:
        TP = len(results_df.loc[(results_df["y_actual"] == value) & (results_df["y_pred"] == value)].index)
        FP = len(results_df.loc[(results_df["y_actual"] != value) & (results_df["y_pred"] == value)].index)
        FN = len(results_df.loc[(results_df["y_actual"] == value) & (results_df["y_pred"] != value)].index)
        TN = len(results_df.loc[(results_df["y_actual"] != value) & (results_df["y_pred"] != value)].index)
        total = TP + TN + FN + FP
        acc = (TP + TN ) / total 
        print("Subject index", str(value+1) + ":", "TP:", TP, "TN:", TN, "FP:", FP, "FN:", FN)
        print("accuracy:", "{:.1%}".format(acc))
    
        
def print_task_2_report(y_actual, y_pred):
    accuracy = accuracy_score(y_actual, y_pred)
    #report combined classification
    print("Combined classification accuracy: " + "{:.1%}".format(accuracy))
    print()
    print_accuracy_by_subject(y_actual, y_pred)
    return accuracy
        
def print_cross_validation_report(cross_validation_list):
    
    def print_stage(index, stage):
        y_actual = stage[0]
        y_pred = stage[1]
        print("Fold " + str(index+1) + ":",)
        accuracy = accuracy_score(y_actual, y_pred)
        #report combined classification
        print("Combined classification accuracy:" , "{:.1%}".format(accuracy))
        print_accuracy_by_subject(y_actual, y_pred)
        print()
        return (list(y_actual), list(y_pred))
        
    def print_overall_report(y_actual_list, y_pred_list):
        print("Overall report:")
        y_actual = np.array(y_actual_total)
        y_pred = np.array(y_pred_total)
        accuracy = accuracy_score(y_actual, y_pred)
        #report combined classification
        print("Overall combined classification accuracy: " + "{:.1%}".format(accuracy))
        print()
        print_accuracy_by_subject(y_actual, y_pred)
        return accuracy
        
    print("Cross Validation Report:")
    print()
    y_actual_total = []
    y_pred_total = []
    print("Fold report")
    for index, stage in enumerate(cross_validation_list):
        next_y_actual, next_y_pred = print_stage(index, stage)
        y_actual_total.extend(next_y_actual)
        y_pred_total.extend(next_y_pred)
    accuracy = print_overall_report(y_actual_total, y_pred_total)
    return accuracy

def perform_task_2(images):
    #get the y array
    y = images.get_y_array()
    x = images.cut_pca_array
    #split the data
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
    #perform the knn
    y_pred = perform_knn(x_train, x_test, y_train)
    return y_test, y_pred

def print_comparison_report(original_images, reduced_images):
    
    def make_comparison(scores):
        print("The original images had an accuracy of", "{:.1%}".format(scores[0]))
        print("The reduced images had an accuracy of", "{:.1%}".format(scores[1]))
        percent_diff = (scores[0] - scores[1]) / ((scores[0] + scores[1]) / 2)
        print("The original images were", "{:.1%}".format(percent_diff), "more accurate than the reduced images.")
    
    print("Comparison of basic knn accuracy:")
    basic_scores = [original_images.basic_knn_accuracy, reduced_images.basic_knn_accuracy]
    make_comparison(basic_scores)
    print()
    print("Comparison of cross validation accuracy:")
    fold_scores = [original_images.cross_validation_accuracy, reduced_images.cross_validation_accuracy]
    make_comparison(fold_scores)
```
